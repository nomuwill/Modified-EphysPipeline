```python
import braingeneers.utils.s3wrangler as wr
import numpy as np
```

```python
uuid_list = [
    "2024-07-05-e-all_cell_lines_dorsal_pasca",
    "2024-07-03-e-sakura-drugs-21985",
    "2024-07-02-e-sakura-drugs-20174a",
    "2024-07-02-e-all_cell_lines_dorsal_pasca",
    "2024-07-01-e-sakura-drugs-22717",
    "2024-06-28-e-all_cell_lines_dorsal_pasca",
    "2024-06-24-e-sakura-drugs-22097",
    "2024-02-09-e-pharma1-20380-HS",
    "2024-04-04-e-p001384-pharma-reupload"
]
```

```python
base_bucket = "s3://braingeneers/ephys"
data_bucket = "original/data"
conn_bucket = "derived/connectivity"

for uuid in uuid_list:
    recordings = wr.list_objects(f"{base_bucket}/{uuid}/{data_bucket}/")
    connectivity = wr.list_objects(f"{base_bucket}/{uuid}/{conn_bucket}/")
    print(f"UUID: {uuid}")
    print(f"Recordings: {len(recordings)}")
    print(f"Connectivity: {len(connectivity)}")
    if len(recordings) > len(connectivity):
        print("Missing connectivity:")
        for rec in recordings:
            file_name = rec.split("/")[-1].split(".")[0]
            for conn in connectivity:
                if file_name in conn:
                    break
            else:
                print(f"  {rec}")
                
```
